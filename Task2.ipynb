```python
# Task2
# Perform data cleaning and exploratory data analysis (EDA) on a dataset of your choice, such as the Titanic dataset from Kaggle. 
# Explore the relationships between variables and identify patterns and trends in the data

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

df2 = pd.read_csv("Titanic.csv")
df2.head()
```

```python
print("Check Missing Values")
df2.isnull().sum()
```

```python
print("Fill missing value in Age Column with Median")
df2['Age'] = df2['Age'].fillna(df2['Age'].median())

print("Fill missing value in Fare Column with Median")
df2['Fare'] = df2['Fare'].fillna(df2['Fare'].median())

print("Drop cabin column in high missing values")
df2.drop(['Cabin'], axis=1, inplace=True)

print("After cleaning check missing values")
df2.isnull().sum()
```

```python
print("Survival count")
sns.countplot(data=df2, x='Survived')
plt.title("Survival Count")
plt.show()
```

```python
print("Survival by Sex")
sns.countplot(data=df2, x='Sex', hue='Survived')
plt.title("Survival by Gender")
plt.show()
```

```python
print("Survival by Class")
sns.countplot(data=df2, x='Pclass', hue='Survived')
plt.title("Survival by Passenger Class")
plt.show()
```

```python
print("Age distribution")
plt.hist(df2['Age'], bins=20, edgecolor='black')
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
print("Age vs Survival")
sns.boxplot(x='Survived', y='Age', data=df2)
plt.title("Age vs Survival")
plt.show()
```
